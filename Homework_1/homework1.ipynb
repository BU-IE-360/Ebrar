### Model of the Unemployment Rate
In today's world, people mostly look for a job in the Internet. Linkedin is one of the global websites which helps people to get a job in different ways. Also, iskur has such a property in Turkey. It's an organization to help people finding suitable jobs and employer finding suitable employees. Along with these two, "is ilanı (job ads in English)" can be a critical key word while looking for a job. Thus, the first model looks for whether there is a relationship between unemployment rate and LinkedIn, Iskur, and job ads or not.

```r
# All trends data will be used following models. It'll be explained part by part, but for now, let's see all data sets together

df2 <- read.csv("trends_last_new.csv")
head(df2)
tail(df2)
```

```r
summary(df2)
str(df2)
```

```r
# Because Google Trends data sets are between 0 and 100, let's normalize bank nata
df$issizlik <- 100*((df$issizlik -min(df$issizlik))/(max(df$issizlik)-min(df$issizlik))) #unemployment rate
df$maddi.beklenti <- 100*((df$maddi.beklenti-min(df$maddi.beklenti))/(max(df$maddi.beklenti)-min(df$maddi.beklenti))) #household's financial expectation
df$eglence.ve.kultur <- 100*((df$eglence.ve.kultur-min(df$eglence.ve.kultur))/(max(df$eglence.ve.kultur)-min(df$eglence.ve.kultur))) #entertainment and culture services
head(df)
tail(df)
```

```r
#Adding trends data to the main data frame
df$linkedin <- df2$linkedin
df$iskur <- df2$iskur
df$is.ilani <- df2$is.ilani
df$sinema <- df2$sinema
df$tiyatro <- df2$tiyatro
df$konser <- df2$konser
df$iddaa <- df2$iddaa
df$sayisal.loto <- df2$sayisal.loto
df$at.yarisi <- df2$at.yarisi
head(df)
summary(df)
str(df)
```

```r
num.cols2 <- sapply(df, is.numeric)
cor.data2 <- cor(df[,num.cols2]) #Calculate correlation
print(cor.data2)
print(corrplot(cor.data2,method="color", tl.cex = 1.1, cl.cex=0.7))
```

```r
# To see all relations as a whole
ggpairs(df, cardinality_threshold = NULL, columns = 2:13)
```

```r
#The first model of the unemployment rate
model_is <- lm(issizlik ~ linkedin+iskur+is.ilani, data=df)
summary(model_is)
```

The first model's adjusted R-squared value is 0.4515. It is not a bad starting point, but not that good either. Looking at probability values, it seems iskur and is.ilani have an important effect in the model rather than LinkedIn. Let's check pairwise relations to make improvement.

```r
# Plot all pairs to see relations
ggplot(df, aes(x=linkedin,y=issizlik)) + geom_point()
ggplot(df, aes(x=iskur,y=issizlik)) + geom_point()
ggplot(df, aes(x=is.ilani,y=issizlik)) + geom_point()
```

```r
# Plot unemployment rate to see its condition
ggplot(df,aes(x=issizlik)) + geom_histogram(fill="#87CEFA")
ggplot(df, aes(x=Tarih, y=issizlik)) + geom_point()
```

LinkedIn seems to have a nonlinear relationship. Firstly, let's try adding a logarithmic term and see whether it improves the model or not.

```r
df_copy1 <- data.frame(df)
head(df_copy1)
```

```r
#The second model of the unemployment rate
df_copy1$link <- log(df_copy1$linkedin)
model_is_new <- lm(issizlik ~ linkedin+iskur+is.ilani+link, data=df_copy1)
summary(model_is_new)
```

```r
checkresiduals(model_is_new)
```

Adjusted R-squared increased by 0.02. It's not that much improvement. Also, there's high autocorrelation in the residuals which distorts the assumptions.
There's no additional clear nonlinear relationship between dependent variable and independent variables. Check autocorrelation in dependent variable, the problem may be a result of it.

```r
acf(df$issizlik)
```

Unemployment data seems autocorrelated. Let's take lag 1 to break this.

```r
#The third model of the unemployment rate
mean <- mean(df_copy1$issizlik)
dif <- c(mean, df_copy1$issizlik[1:120])
model_is_new1 <- lm(issizlik ~ linkedin+iskur+is.ilani+link+dif, data=df_copy1)
summary(model_is_new1)
```

```r
plot(model_is_new1)
```

The model improved a lot. Let's check whether assumptions hold in the residual side or not.

```r
checkresiduals((model_is_new1))
```

At some points, residuals have correlation more than desired (lag 6, lag 12, lag 18). However, most of them seem okay, below or near the level. Residuals seem to distribute normally, so that assumption holds, too.
Taking a lag 6 may decrease that autocorrelation, but because 5 terms are already added to the model, I will not continue with this.

### The Model of Advertisement and Culture Services Price Index
Going to cinema (sinema in Turkish), theatre (tiyatro), and concert(konser) are some fundamental fun activities. They are reachable for most of the population, so there may be a relation between them and advertisement and culture services price index. This model searches for whether there is a relation between advertisement-culture services price index (eglence.ve.kultur in the model) and cinema, theatre, concert or not.

```r
# The first model of the advertisement and culture services price index
model_eglence <- lm(eglence.ve.kultur ~ sinema+tiyatro+konser, data=df)
summary(model_eglence)
checkresiduals(model_eglence)
```

The starting model's adjusted R-squared value is 0.3978. It's worse than issizlik case. Let's check plot of pairs to improve the model.

```r
ggplot(df, aes(x=sinema,y=eglence.ve.kultur)) + geom_point()
ggplot(df, aes(x=tiyatro,y=eglence.ve.kultur)) + geom_point()
ggplot(df, aes(x=konser,y=eglence.ve.kultur)) + geom_point()
```

```r
# To see dependent variable by itself
ggplot(df,aes(x=eglence.ve.kultur)) + geom_histogram(fill="#87CEFA")
ggplot(df, aes(x=Tarih, y=eglence.ve.kultur)) + geom_point()
```

Advertisement and culture services price index (eglence.ve.kultur) seems to distribute exponentially. Taking its logarithm may help with this.

```r
df_copy2 <- data.frame(df)
```

```r
# The second model of the advertisement and culture services price index
model_eglence_new <- lm(log(eglence.ve.kultur + 1)~ sinema+tiyatro+konser, data=df_copy2)
summary(model_eglence_new)
checkresiduals(model_eglence_new)
```

The adjusted R-squared value improved by nearly 0.14. However, there's autocorrelation between residuals, and they seem not to have stationary mean. Model needs further improvement.
In the plots, cinema seems to have a nonlinear relation with eglence.ve.kultur. Let's try adding logarithm with it too.

```r
# The third model of the advertisement and culture services price index
df_copy2$lg_sinema <- log(df_copy2$sinema)
model_eglence_new1 <- lm(log(eglence.ve.kultur + 1)~ sinema+tiyatro+konser+lg_sinema, data=df_copy2)
summary(model_eglence_new1)
checkresiduals(model_eglence_new1)
```

Adjusted R-squared improved a bit in this case, too, but not as it is wanted. Another approach should be tried. After taking the logarithm of a exponential distribution, it will look like a trend line. Therefore adding a trend can also improve this model. Let's try this.

```r
# To see the trend
ggplot(df, aes(x=Tarih, y=log(eglence.ve.kultur + 1))) + geom_point()
```

```r
# The fourth model of the advertisement and culture services price index
df_copy2$trend <- 1:121
model_eglence_new2 <- lm(log(eglence.ve.kultur + 1) ~ sinema+tiyatro+konser+lg_sinema+trend, data=df_copy2)
summary(model_eglence_new2)
```

Trend, logarithm coefficient, and sinema affect more than konser and tiyatro.

```r
plot(model_eglence_new2)
```

Adjusted R-squared improved a lot. However, it is not enough. Let's check residuals to see whether the assumptions hold or not.

```r
checkresiduals(model_eglence_new2)
```

Although it has a high adjusted R-squared value, this model does not hold main assumptions. There seems high correlation between residuals. Also, there's a pick which distorts the normality assumption. Looking at all steps, it seems that there is high error after 100th data. Excluding it, and redoing the operations may give better results. (It may be due to covid period or country's economic condition) Because I tried adding 5 variables, I stopped here for this one. However, I will try another approach for this to see if I can have residuals which hold the assumptions.

```r
acf(df_copy2$eglence.ve.kultur)
```

Data seems to have autocorrelation. Try lag 1 to fix this.

```r
# The alternative model
dif1 <- c(0, df_copy2$eglence.ve.kultur[1:120])
model_eglence_new3 <- lm(eglence.ve.kultur ~ sinema+tiyatro+konser+lg_sinema+dif1, data=df_copy2)
summary(model_eglence_new3)
checkresiduals(model_eglence_new3)
```

```r
plot(model_eglence_new3)
```

Lag, sinema and tiyatro has high affects. Also, residuals after 100th data point (nearly 110th) still creates problem. However, autocorrelation between residuals are much lower in this case. Lag 2,10,12 has higher values than desired, but as a whole, it is better in the residual side.

### The Model of Financial Expectation of the Household
People tend to play games on chance when they are not in a good situation economically. Games on chance are seen as an easy way of earning money, so they play such games to improve their financial situation. Some examples of these games are iddaa which is Turkey's official football betting game, sayisal loto (policy? not sure of the translation), and at yarisi (horse racing/derby). This model tries to understand whether there is a relation between financial expectation of the household and iddaa, sayisal loto, at yarisi in reality.

```r
# The first model of financial expectation
model_maddi <- lm(maddi.beklenti ~ iddaa+at.yarisi+sayisal.loto, data=df)
summary(model_maddi)
```

The starting model has 0.441 adjusted R-squared value. "iddaa" and "at yarisi" seem to affect financial expectation, but "sayisal.loto" does not have much power. Let's check pairwise relations and the relation in the dependent data to improve the model.

```r
ggplot(df, aes(x=iddaa,y=maddi.beklenti)) + geom_point()
ggplot(df, aes(x=sayisal.loto,y=maddi.beklenti)) + geom_point()
ggplot(df, aes(x=at.yarisi,y=maddi.beklenti)) + geom_point()
```

```r
ggplot(df,aes(x=maddi.beklenti)) + geom_histogram(fill="#87CEFA")
ggplot(df, aes(x=Tarih, y=maddi.beklenti)) + geom_point()
```

"iddaa" seems to have a nonlinear relation with "maddi.beklenti" (economic expectation). Let's try this.

```r
df_copy3 <- data.frame(df)
```

```r
# The second model of financial expectation
df_copy3$id <- sqrt(df_copy3$iddaa)
model_maddi_new <- lm(maddi.beklenti~ iddaa+sayisal.loto+at.yarisi+id, data=df_copy3)
summary(model_maddi_new)
```

```r
checkresiduals(model_maddi_new)
```

Adjusted R-squared value improved nearly 0.1. However, residuals seem to have an increasing variance. There's no nonlinear relation between dependent and independent variables which can be seen directly. Let's check autocorrelation within dependent variable.

```r
acf(df_copy3$maddi.beklenti)
```

"maddi.beklenti" (financial expectation) data seems to have autocorrelation within itself. Let's try lag 1 to fix this.

```r
# The third model of financial expectation
mean3 <- mean(df_copy3$maddi.beklenti)
dif2 <- c(mean3, df_copy3$maddi.beklenti[1:120])
model_maddi_new1 <- lm(maddi.beklenti~ iddaa+sayisal.loto+at.yarisi+id+dif2, data=df_copy3)
summary(model_maddi_new1)
```

Adjusted R-squared improved. "iddaa" and "at yarisi" are still important, but lag has a higher affect. Let's check residuals.

```r
checkresiduals(model_maddi_new1)
h <- c(0, diff(df_copy3$maddi.beklenti, lag=1))
ggplot(df, aes(x=Tarih, y=h)) + geom_point()
```

```r
plot(model_maddi_new1)
```

Residuals do not have autocorrelation between themselves. Their increasing variance seems to decrease, but it is not fixed totally. There seems a distortion after 110th data. It may result from covid of economic conditions (?). It needs further research. Other methods can be tried to fix this. Because I added 5 variables, I stopped here.
